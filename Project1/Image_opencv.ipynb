```python
import cv2
import matplotlib.pyplot as plt

```

```python
# Function to process an image and detect edges
def image_edge_detection(image_path):
    # Read the image
    image = cv2.imread(image_path, 1)
    
    # Apply Gaussian Blur,Median and Bialetral  to smooth the image
    gaussianblurred_image = cv2.GaussianBlur(image, (7, 7), 0)
    medianblurred_image = cv2.medianBlur(image, 5)
    bilateralblurred_image = cv2.bilateralFilter(image, 9, 75, 75)
    
    # Perform Canny Edge Detection
    g_edges = cv2.Canny(gaussianblurred_image, threshold1=50, threshold2=120)
    m_edges = cv2.Canny(medianblurred_image, threshold1=50, threshold2=120)
    b_edges = cv2.Canny(bilateralblurred_image, threshold1=50, threshold2=120)
    
    # Display the original and edge-detected images
    plt.figure(figsize=(15, 6))
    
    # Original Image
    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    # Edge-Detected Image1
    plt.subplot(1, 4, 2)
    plt.imshow(g_edges, cmap='gray')
    plt.title('Gaussian Blur Detection')
    plt.axis('off')
    # Edge-Detected Image2
    plt.subplot(1, 4, 3)
    plt.imshow(m_edges, cmap='gray')
    plt.title('Median Blur Detection')
    plt.axis('off')
    # Edge-Detected Image3
    plt.subplot(1, 4, 4)
    plt.imshow(m_edges, cmap='gray')
    plt.title('Bialateral Blurr Detection')
    plt.axis('off')
    
    # Show the plots
    plt.show()

```

```python
# Path to the image file
image_path = 'Pic2.jpg'  # Replace with your image path
image_edge_detection(image_path)
```
